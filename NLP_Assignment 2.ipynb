```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = text.lower()
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    tokens = [w for w in tokens if len(w) > 2 or w in ["no", "not"]]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for s in sentences:
    tokens, clean = preprocess_text(s)

    print("\n===================")
    print("Original:", s)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
```

```python
def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg = sum(len(w) for w in tokens) / total if total > 0 else 0

    return total, unique, round(avg, 2)
```

```python
for s in sentences:
    tokens, _ = preprocess_text(s)
    t, u, a = token_analysis(tokens)

    print("\nSentence:", s)
    print("Total:", t)
    print("Unique:", u)
    print("Avg Length:", a)
```

```python
all_tokens = []

for s in sentences:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10:", counter.most_common(10))
print("Least 5:", counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(sentences)
print(result)
```

```python
edge_cases = ["", "😍😍😍", "123456"]

for case in edge_cases:
    tokens, clean = preprocess_text(case)

    print("\nCase:", case)
    print("Tokens:", tokens)
    print("Clean:", clean)
```
